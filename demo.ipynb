```python
from PIL import Image
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from invert import Invert
```

```python
inv = Invert()

gr = Image.open('img/gr.png')     # MNIST
rgb = Image.open('img/rgb.png')   # CIFAR10
rgba = Image.open('img/rgba.png') # CIFAR10 with a transparent hole in the middle
```

```python
plt.figure()
plt.imshow(np.asarray(gr), cmap='gray')
plt.figure()
plt.imshow(np.asarray(inv(gr)), cmap='gray')
```

```python
plt.figure()
plt.imshow(np.asarray(rgb))
plt.figure()
plt.imshow(np.asarray(inv(rgb)))
```

```python
plt.figure()
plt.imshow(np.asarray(rgba))
plt.figure()
plt.imshow(np.asarray(inv(rgba)))
```
